# Chi-Square

### 1. Does the choice of Ground significantly impact the probability of Team 1 winning?

```python
contingency_table = pd.crosstab(df['Ground'], df['Winner'])

chi2, p, _, _ = stats.chi2_contingency(contingency_table)

#The null hypothesis is that the choice of 
#ground has no impact on the probability of Team 1 winning. 

#significance level (alpha)
alpha = 0.05

# Checking if the p-value is less than alpha
if p < alpha:
    print("There is a significant association between Ground and Team 1 winning.")
else:
    print("Ground choice does not significantly impact the probability of Team 1 winning.")
```

# 2. Is there a correlation between the Winner and the Ground?

```python
from sklearn.preprocessing import LabelEncoder 

label_encoder = LabelEncoder()
df['Winner_encoded'] = label_encoder.fit_transform(df['Winner'])
df['Ground_encoded'] = label_encoder.fit_transform(df['Ground'])

# contingency table
contingency_table = pd.crosstab(df['Winner_encoded'], df['Ground_encoded'])

# chi-square test
chi2, p, _, _ = stats.chi2_contingency(contingency_table)

# significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha
if p < alpha:
    print("There is a significant correlation between Winner and Ground.")
else:
    print("There is no significant correlation between Winner and Ground.")
```

### Since the p-value is less than alpha, we reject the null hypothesis and conclude that there is a significant correlation between the Winner and Ground columns.

# ANOVA

# 3.Can we conclude that the Home or Away status of the teams affects the match results?

```python

# ANOVA test
f_stat, p_value = stats.f_oneway(df[df['Host_Country'] == 'Home']['Winner'], df[df['Host_Country'] == 'Away']['Winner'])

#significance level (alpha)
alpha = 0.05

# Checking if the p-value is less than alpha
if p_value < alpha:
    print("We can conclude that the Home or Away status of the teams significantly affects the match results.")
else:
    print("We cannot conclude that the Home or Away status of the teams significantly affects the match results.")
```

### This suggests that the home/away factor does affect the outcomes of matches.

# F-test

# 4. Does the Winner's identity depend on whether they played at Home or Away?

```python
# ANOVA F-test
f_stat, p_value = stats.f_oneway(df[df['Host_Country'] == 'Home']['Winner'], df[df['Host_Country'] == 'Away']['Winner'])

# Set a significance level (alpha)
alpha = 0.05

# Checking if the p-value is less than alpha
if p_value < alpha:
    print("The Winner's identity depends on whether they played at Home or Away.")
else:
    print("The Winner's identity does not significantly depend on whether they played at Home or Away.")
```

### The p-value is greater than alpha,The Winner's identity does not significantly depend on whether they played at Home or Away.

# T-test

# 5. Do teams that bat first (Innings_Team1) tend to score more runs on average compared to teams that bat second?

```python
df['Margin'] = pd.to_numeric(df['Margin'], errors='coerce')

team1_scores = df[df['Innings_Team1'] == 'First']['Margin']
team2_scores = df[df['Innings_Team2'] == 'First']['Margin']


team1_scores = team1_scores.dropna()
team2_scores = team2_scores.dropna()

#two-sample t-test
t_stat, p_value = stats.ttest_ind(team1_scores, team2_scores, equal_var=False)

alpha = 0.05

# Check if the p-value is less than alpha
if p_value < alpha:
    print("Teams batting first tend to score significantly more runs on average.")
else:
    print("There is no significant difference in the average scores between teams batting first and second.")
```

### The hypothesis test conducted did not find a significant difference in the average scores of teams based on whether they batted first or second.
